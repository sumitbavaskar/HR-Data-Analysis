# logistic regression


```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report

# Scale the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_smote)
X_test_scaled = scaler.transform(X_test)

# Logistic Regression with stable solver
model = LogisticRegression(solver='liblinear', max_iter=1000)
model.fit(X_train_scaled, y_train_smote)

# Predict and evaluate
y_pred = model.predict(X_test_scaled)
print(" Accuracy:", accuracy_score(y_test, y_pred))
print(" Report:\n", classification_report(y_test, y_pred))
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report

# Train KNN with k=5 (default)
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scaled, y_train_smote)

# Predict on test data
y_pred_knn = knn.predict(X_test_scaled)

# Evaluate
print(" KNN Accuracy:", accuracy_score(y_test, y_pred_knn))
print(" Classification Report:\n", classification_report(y_test, y_pred_knn))
```

```python
error_rates = []
from sklearn.metrics import accuracy_score

for k in range(1, 21):
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train_scaled, y_train_smote)
    preds = model.predict(X_test_scaled)
    error = 1 - accuracy_score(y_test, preds)
    error_rates.append(error)

error_rates
```

```python
# Plot Error Rate vs. K
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(range(1, 21), error_rates, color='red', linestyle='dashed', marker='o')
plt.title("Error Rate vs. K Value")
plt.xlabel("K")
plt.ylabel("Error Rate (1 - Accuracy)")
plt.xticks(range(1, 21))
plt.grid(True)
plt.show()
```
